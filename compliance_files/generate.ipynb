```python
def union_of_lists(lists):
    return list(set(item for sublist in lists for item in sublist))
```

```python

def convert_str_to_list(str_list):
    import ast
    return ast.literal_eval(str_list)
```

```python
regulation = dataframe['refer_regulation']
regulations_lst = []
for i in regulation:
    regulations_lst.append(convert_str_to_list(i))
```

```python
print(regulations_lst)
a = (list(item for sublist in regulations_lst for item in sublist))
print(set(a))
```

```python
def merge_dataframe_columns(df, columns, new_column_name):
    df[new_column_name] = df[columns].astype(str).agg(' '.join, axis=1)
    return df
```
